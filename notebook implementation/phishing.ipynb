```python
plt.figure(figsize=(15,500), facecolor='white')
plotnumber = 1

for column in data.drop(['Result'],axis=1):
    ax = plt.subplot(12, 3, plotnumber)
    sns.countplot(data[column])
    plt.xlabel(column, fontsize=30)
    plotnumber += 1
plt.show()
```
